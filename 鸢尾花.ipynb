```python
import os
os.environ['OMP_NUM_THREADS'] = '1'

from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt

# 加载鸢尾花数据集
iris = load_iris()

# 创建 DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# 提取特征数据
x = iris.data[:, :4]

# %%
from sklearn.cluster import KMeans

# 创建 KMeans 聚类器
estimator = KMeans(n_clusters=3)  # 聚类数设为 3
estimator.fit(x)  # 拟合模型
label = estimator.labels_  # 获取聚类标签

# 提取每个聚类的数据点
x0 = x[label == 0]  # 聚类 0
x1 = x[label == 1]  # 聚类 1
x2 = x[label == 2]  # 聚类 2

# %%
# 绘制二维散点图
plt.scatter(x0[:, 0], x0[:, 1], c="red", label='Setosa')  # 聚类 0 用红色表示
plt.scatter(x1[:, 0], x1[:, 1], c="blue", label='Versicolour')  # 聚类 1 用蓝色表示
plt.scatter(x2[:, 0], x2[:, 1], c="green", label='Virginica')  # 聚类 2 用绿色表示
plt.xlabel(iris.feature_names[0])  # x 轴标签
plt.ylabel(iris.feature_names[1])  # y 轴标签
plt.legend(loc=2)  # 在左上角显示图例
plt.title("2D Scatter Plot of Iris Clusters")
plt.show()  # 显示图形

# %%
# 绘制三维散点图
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')  # 创建 3D 坐标轴

# 在 3D 空间中绘制每个聚类的散点图
ax.scatter(x0[:, 0], x0[:, 1], x0[:, 2], c="red", label='Setosa')
ax.scatter(x1[:, 0], x1[:, 1], x1[:, 2], c="blue", label='Versicolour')
ax.scatter(x2[:, 0], x2[:, 1], x2[:, 2], c="green", label='Virginica')

# 设置坐标轴标签
ax.set_xlabel(iris.feature_names[0])  # x 轴标签
ax.set_ylabel(iris.feature_names[1])  # y 轴标签
ax.set_zlabel(iris.feature_names[2])  # z 轴标签
ax.legend(loc=2)  # 在左上角显示图例
plt.title("3D Scatter Plot of Iris Clusters")
plt.show()  # 显示图形

# %%
# 绘制四维数据的 3D 散点图
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# 将数据分为四个特征
x = iris.data[:, :1]  # 第一个特征作为 x 轴
y = iris.data[:, 1:2]  # 第二个特征作为 y 轴
z = iris.data[:, 2:3]  # 第三个特征作为 z 轴
c = iris.data[:, 3]  # 第四个特征用于颜色编码

# 创建 3D 散点图，使用 'hot' 色图表示第四个特征
img = ax.scatter(x, y, z, c=c, cmap=plt.hot())
plt.colorbar(img)  # 添加颜色条，以指示第四个特征的值
ax.set_xlabel(iris.feature_names[0])  # x 轴标签
ax.set_ylabel(iris.feature_names[1])  # y 轴标签
ax.set_zlabel(iris.feature_names[2])  # z 轴标签
plt.title("4D Scatter Plot of Iris Data")
plt.show()  # 显示图形
```
