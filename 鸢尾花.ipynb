```python
import os
os.environ['OMP_NUM_THREADS'] = '1'

from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt

# 
iris = load_iris()

#  DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# 
x = iris.data[:, :4]

# %%
from sklearn.cluster import KMeans

#  KMeans 
estimator = KMeans(n_clusters=3)  #  3
estimator.fit(x)  # 
label = estimator.labels_  # 

# 
x0 = x[label == 0]  #  0
x1 = x[label == 1]  #  1
x2 = x[label == 2]  #  2

# %%
# 
plt.scatter(x0[:, 0], x0[:, 1], c="red", label='Setosa')  #  0 
plt.scatter(x1[:, 0], x1[:, 1], c="blue", label='Versicolour')  #  1 
plt.scatter(x2[:, 0], x2[:, 1], c="green", label='Virginica')  #  2 
plt.xlabel(iris.feature_names[0])  # x 
plt.ylabel(iris.feature_names[1])  # y 
plt.legend(loc=2)  # 
plt.title("2D Scatter Plot of Iris Clusters")
plt.show()  # 

# %%
# 
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')  #  3D 

#  3D 
ax.scatter(x0[:, 0], x0[:, 1], x0[:, 2], c="red", label='Setosa')
ax.scatter(x1[:, 0], x1[:, 1], x1[:, 2], c="blue", label='Versicolour')
ax.scatter(x2[:, 0], x2[:, 1], x2[:, 2], c="green", label='Virginica')

# 
ax.set_xlabel(iris.feature_names[0])  # x 
ax.set_ylabel(iris.feature_names[1])  # y 
ax.set_zlabel(iris.feature_names[2])  # z 
ax.legend(loc=2)  # 
plt.title("3D Scatter Plot of Iris Clusters")
plt.show()  # 

# %%
#  3D 
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# 
x = iris.data[:, :1]  #  x 
y = iris.data[:, 1:2]  #  y 
z = iris.data[:, 2:3]  #  z 
c = iris.data[:, 3]  # 

#  3D  'hot' 
img = ax.scatter(x, y, z, c=c, cmap=plt.hot())
plt.colorbar(img)  # 
ax.set_xlabel(iris.feature_names[0])  # x 
ax.set_ylabel(iris.feature_names[1])  # y 
ax.set_zlabel(iris.feature_names[2])  # z 
plt.title("4D Scatter Plot of Iris Data")
plt.show()  # 
```
